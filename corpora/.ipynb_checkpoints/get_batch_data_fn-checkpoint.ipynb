```python
def load_source_vocab():
    return load_vocab('voca.src')


def load_target_vocab():
    return load_vocab('voca.tgt')
```

```python
def load_vocab(path):
    vocab = [line.rstrip() for line in codecs.open(path, 'r', 'utf-8')]
    word2idx = {word: idx for idx, word in enumerate(vocab)}
    idx2word = {idx: word for idx, word in enumerate(vocab)}
    return word2idx, idx2word
```

```python
src_sents = [line for line in
                 codecs.open('train.src', 'r', 'utf-8').read().split("\n") if line]
tgt_sents = [line for line in
                 codecs.open('train.tgt', 'r', 'utf-8').read().split("\n") if line]
```

```python
src_sents[14]
```

```python
x_list, y_list = [], []
```

```python
from tqdm import tqdm
import re
import math
import numpy as np
```

```python
_WORD_SPLIT = re.compile("([.,!?\"':;)(])")

def basic_tokenizer(sentence, lower=True):
    words = []
    for space_separated_fragment in sentence.strip().split():
        words.extend(_WORD_SPLIT.split(space_separated_fragment))
    return [w.lower() if lower else w for w in words if w != '' and w != ' ']
```

```python
for source_sent, target_sent in tqdm(zip(src_sents, tgt_sents), desc="Preparing data: ", total=len(src_sents)):
        x = [src2idx.get(word, src2idx["<unk>"]) for word in basic_tokenizer(source_sent,lower=False)]
        y = [tgt2idx.get(word, tgt2idx["<unk>"]) for word in basic_tokenizer(target_sent)]
        if len(x) != len(y):
            print(x)
            continue
        if len(x) < 5:
            continue
        n_samples = math.ceil(len(x)/30)
        for i in range(n_samples):
            maybe_x = np.array(x[i*30:(i+1)*30])
            if np.sum(maybe_x > 8) < 5:
                continue
            x_list.append(maybe_x)
            y_list.append(np.array(y[i*30:(i+1)*30]))
```

```python
x_list[14]
```

```python
 X = np.zeros([len(x_list), 30], np.int32)
Y = np.zeros([len(y_list), 30], np.int32)
```

```python
X
```

```python
Y
```

```python
for i, (x, y) in tqdm(enumerate(zip(x_list, y_list)), desc="Padding: ", total=len(x_list)):
        X[i] = np.lib.pad(x, [0, 30 - len(x)], 'constant', constant_values=(0, 0))
        Y[i] = np.lib.pad(y, [0, 30 - len(y)], 'constant', constant_values=(0, 0))
```
